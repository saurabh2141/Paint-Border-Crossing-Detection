# Paint Border Crossing Detection

### I have used the provided image from the data for this test and sliced the image to get the bottom left rectangle for Paint Border Crossing Detection.

```python
# importing necessary libraries

import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# importing the image and loading the image

img=cv2.imread('Picture1.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5,5))
plt.imshow(img_rgb)
plt.title('Image')
plt.axis('off')
plt.show()
```

```python
# shape of the imported image

print('Shape of the image:',img.shape)
shape=pd.DataFrame(img.shape)
shape.index=['Height','width','channels']
shape.rename(columns={0:'Values'},inplace=True)
shape
```

```python
# lets slice the above jpg file to display only the bottom left blue rectangular box

image=img[400:550,50:300]
rgb_image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5,5))
plt.imshow(rgb_image)
plt.title('Image')
plt.axis('off')
plt.show()
```

```python
# shape of the sliced image

print('Shape of the sliced image:',image.shape)
shape_new=pd.DataFrame(image.shape)
shape_new.index=['Height','width','channels']
shape_new.rename(columns={0:'Values'},inplace=True)
shape_new
```

```python
# Define the rectangle boundary
x_start, y_start, x_end, y_end = 42, 15, 151, 123

# define blue color range
lower_blue = np.array([100, 150, 0])
upper_blue = np.array([140, 255, 255])

hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

hsv_image_color=cv2.cvtColor(hsv_image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5,5))
plt.imshow(hsv_image_color)
plt.title('Image')
plt.axis('off')
plt.show()
```

```python
# Create a mask for blue color
mask = cv2.inRange(hsv_image, lower_blue, upper_blue)
mask_color=cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5,5))
plt.imshow(mask_color)
plt.title('Image')
plt.axis('off')
plt.show()
```

```python
outside_boundary = mask.copy()

# Set the area inside the rectangle to 0
outside_boundary[y_start:y_end, x_start:x_end] = 0

# Check if there are any blue pixels outside the boundary
if np.any(outside_boundary):
    print("There are blue pixels outside the boundary.")
else:
    print("There are no blue pixels outside the boundary.")
```

```python

outside_boundary_color=cv2.cvtColor(outside_boundary, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5,5))
plt.imshow(outside_boundary_color)
plt.title('Image')
plt.axis('off')
plt.show()
```
